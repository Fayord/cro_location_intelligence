```python
import pandas as pd
import os
```

```python
dir_path = os.getcwd()
result_folder = f"{dir_path}/result/"
# data_path in result_folder
data_name_list = os.listdir(result_folder)
data_name_list.sort()
print(data_name_list)
```

```python
all_df = pd.DataFrame()
for data_name in data_name_list:
    data_path = result_folder + data_name
    # print(data_path)
    target_store_id = data_name.split("/")[-1].split(".")[0]
    target_store_id = int(target_store_id)
    df = pd.read_csv(data_path)
    df["input_store_id"] = target_store_id
    # drop pred column
    df = df.drop(columns=["pred"])
    # rename store_id to pred_store_id
    df = df.rename(columns={"store_id": "pred_store_id"})
    all_df = pd.concat([all_df, df], ignore_index=True)
    display(df)
    break
```

```python
# save all_df
all_df.to_csv(f"{dir_path}/all_df.csv", index=False)
```
