```python
import osmnx as ox
import networkx as nx
from  networkx.classes.multidigraph import MultiDiGraph 
import matplotlib.pyplot as plt
import geopandas as gpd
from typing import List, Tuple, Dict, Set, Optional
```

```python
def get_highway_data(place_name:str)->MultiDiGraph:
    graph = ox.graph_from_place(
        place_name, 
        network_type="drive", # all, all_private, bike, drive, drive_service, walk
        simplify=False, # if turn true it reduce node edges but use more time
        # custom_filter='["highway"~"motorway|trunk"]', # to filter only highway(motorway or trunk)

        )
    return graph

def convert_to_geopandas(graph:MultiDiGraph)->Tuple[gpd.GeoDataFrame, gpd.GeoDataFrame]:
    gdf_nodes, gdf_edges = ox.graph_to_gdfs(graph)
    return gdf_nodes, gdf_edges
```

```python
place_name = "Chonburi,pattaya"
graph = get_highway_data(place_name)
nodes, edges = convert_to_geopandas(graph)
```

```python
sample = nodes.sample(n=100)
sample.explore()
```
