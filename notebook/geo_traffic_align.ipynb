```python
from PIL import Image, ImageDraw, ImageChops


def place_image_over_with_opacity(
    background,
    foreground,
    center_x,
    center_y,
    resize_size,
    opacity,
):
    # Open the background and foreground images
    bg = Image.open(background)
    fg = Image.open(foreground)
    print("bg.size", bg.size)
    # Resize the foreground image to fit the specified coordinates
    print(fg.size)
    resize_ratio = resize_size / bg.size[1]
    print(bg.size[1] * resize_ratio)
    print("resize_ratio", resize_ratio)
    fg = fg.resize((resize_size, resize_size))

    # Create a copy of the background image to avoid modifying the original
    result = bg.copy()

    # Create a mask with the specified opacity
    mask = Image.new("L", (resize_size, resize_size), int(opacity * 255))
    # Paste the foreground image onto the result using the mask
    x = center_x - resize_size // 2
    y = center_y - resize_size // 2
    result.paste(fg, (x, y), mask)

    # Save or show the resulting image
    return result


# Example usage:
background_image_path = traffic_image_path
foreground_image_path = ref_image_path
x_coordinate = 3456 // 2
y_coordinate = 1788 // 2
opacity_value = 0.5

resize_ratio = 1475

new_image = place_image_over_with_opacity(
    background_image_path,
    foreground_image_path,
    x_coordinate,
    y_coordinate,
    resize_ratio,
    opacity_value,
)
print(x_coordinate)
# set plt size
plt.rcParams["figure.figsize"] = (20, 20)
plt.imshow(new_image)
```
